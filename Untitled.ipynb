```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("nltkdata/state-union-corpus")

print("Path to dataset files:", path)
```

```python
speeches_path = '/Users/smc77/.cache/kagglehub/datasets/nltkdata/state-union-corpus/versions/1/state_union/'
import os; 
speeches_files = os.listdir(speeches_path)
```

```python
import os
import re
import io
import zipfile
import requests
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from pathlib import Path


data = []
keywords = ["self-reliance", "selfreliance", "responsibility", "duty", "independence", "initiative"]

for file in speeches_files:
    with open(speeches_path + file, "r", encoding="utf-8", errors="ignore") as f:
        text = f.read()
        match = re.search(r"(\d{4})", f.name)
        if not match:
            continue
        year = int(match.group(1))
        words = re.findall(r'\b\w+\b', text.lower())
        word_count = len(words)
        counts = Counter(words)

        record = {"year": year, "total_words": word_count}
        record["self_reliance"] = counts["self-reliance"] + counts["selfreliance"]
        for kw in keywords[2:]:  # skip first two since already included
            record[kw] = counts[kw]
        data.append(record)

# Step 3: Create DataFrame and normalize
df = pd.DataFrame(data).sort_values("year")
for kw in ["self_reliance", "responsibility", "duty", "independence", "initiative"]:
    df[f"{kw}_per10k"] = df[kw] / df["total_words"] * 10000

# Step 4: Plot trends
plt.figure(figsize=(12, 6))
for kw in ["self_reliance", "responsibility", "duty", "independence", "initiative"]:
    plt.plot(df["year"], df[f"{kw}_per10k"], label=kw.replace("_", " ").capitalize())

plt.title("Mentions of Self-Reliance Themes in U.S. State of the Union Speeches")
plt.xlabel("Year")
plt.ylabel("Mentions per 10,000 Words")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
